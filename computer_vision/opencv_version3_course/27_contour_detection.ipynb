# Contour Detection

Contours are defined as simply a curve joining all the continuous points (along the boundary), having some colour intensity. Contours are useful for shape analysis and object detection and recognition

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
# Convenience functions 

def show(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)

def show_as_grayscale(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = cv2.imread('DATA/internal_external.png', 0)
show_as_grayscale(img)
```

```python
image, contours, hierarchy = cv2.findContours(img, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
```

```python
ex_contours = np.zeros(image.shape)
```

```python
for i in range(len(contours)):
    # external contour
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(ex_contours, contours, i, 255, -1)
        
```

```python
show_as_grayscale(ex_contours)
```

```python
int_contours = np.zeros(image.shape)

for i in range(len(contours)):
    # internal contour
    if hierarchy[0][i][3] != -1:
        cv2.drawContours(int_contours, contours, i, 255, -1)
```

```python
show_as_grayscale(int_contours)
```
