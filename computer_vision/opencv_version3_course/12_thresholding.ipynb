# Thresholding

In openCV application it is often a practice to convert comlor images to grayscale because it typically will be edges and shapes that are the most important. The color is less important in these cases. Some applications require ony a binary image showing general shapes.

Thresholding is a method of segmenting images to composite or different parts. Thresholding will convert an image to consist of only two parts, black & white. 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Load Image Data

```python
img = cv2.imread('DATA/rainbow.jpg')
plt.imshow(img)
print(img.shape)
```

* Tip - if you want to read in a color image in grayscale you can use an additional parameter to do so. 

```python
img = cv2.imread('DATA/rainbow.jpg', 0)
plt.imshow(img, cmap='gray')
print(img.shape)
```

```python
# threshold
ret, thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
```

```python
plt.imshow(thresh1, cmap='gray')
print("Threshold was: ", ret)
```

### Typical threshold values

- THRESH_BINARY - dictates a black or a white when evaluated against the threshold. All values are min, or max.
- THRESH_BINARY_INV - reverse of the above. 
- THRESH_TRUNC - truncates values above the threshold to be set to the threshold value, others at or below are untouched
- THRESH_TOZERO - keep the source is greater than the threshold, everything below it is set to zero

# Working Example

```python
img = cv2.imread('DATA/crossword.jpg', 0)
plt.imshow(img, cmap='gray')
print(img.shape)
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
show_pic(img)
```

```python
ret, th1 = cv2.threshold(img, 160, 255, cv2.THRESH_BINARY)
show_pic(th1)
```

### Adaptive threshold

```python
adaptive thresholding looks at neighbourhoods of pixels
block size defines the neighbourhood pixels region (needs to be an odd number)
```

```python
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 5, 30)
show_pic(th2)
```

```python
# we can also blend 
blended = cv2.addWeighted(src1=th1, alpha=0.5, src2=th2, beta=0.5, gamma=0)
show_pic(blended)
print(blended.shape)
```
