```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('dark_background')
```

```python
blank_img = np.zeros(shape=(512,512,3), dtype=np.int16)
```

```python
blank_img.shape
```

## Create a rectangle on a blank image

```python
# plotting a rectangle will take two points, top left and bottom right, the order
# of application is along the x-axis, down the y-axis, so remembering that the rule
# is horizontal axis first and top to bottom for vertical axis. 
rec = cv2.rectangle(blank_img, pt1=(383,20), pt2=(500,160), color=(0, 255, 0), thickness=4)
```

```python
plt.imshow(rec)
```

```python
next_rec = cv2.rectangle(blank_img, pt1=(200,200), pt2=(300,300), color=(255,0,0), thickness=5)
```

```python
plt.imshow(next_rec)
```

## Change shape type to Circles

```python
circ = cv2.circle(img=blank_img, center=(100,100), radius=50, color=(255,255,0), thickness=3)
plt.imshow(circ)
```

## Using a negative thickness to obtain a fill

```python
circ = cv2.circle(img=blank_img, center=(300,400), radius=50, color=(255,150,0), thickness=-1)
plt.imshow(circ)
```

## Drawing lines 

```python
liner = cv2.line(blank_img, pt1=(100,100), pt2=(400,400), color=(100,175, 250), thickness=5)
plt.imshow(liner)
```

## fonts 

```python
font = cv2.FONT_HERSHEY_SIMPLEX
textdemo = cv2.putText(blank_img, text="Text demo", org=(20, 490), fontFace=font, fontScale=2, color=(255,255,0), thickness = 2, lineType=cv2.LINE_AA)
plt.imshow(textdemo)
```

# Working with custom polygon

```python
blank_img = np.zeros(shape=(512,512,3), dtype=np.int32)
plt.imshow(blank_img)
```

```python
vertices = np.array([ [100,300], [200,200], [400,300], [200,400] ], np.int32)
```

```python
# standard reshaping to transform the struct to 
# 3d and allow the color channels to make sense
pts = vertices.reshape((-1,1,2))
```

```python
cv2.polylines(blank_img, [pts], isClosed=True, color=(255,0,0), thickness=2)
plt.imshow(blank_img)
```
