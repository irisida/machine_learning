#### TASK: Open the *dog_backpack.jpg* image from the DATA folder and display it in the notebook. Make sure to correct for the RGB order.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("dark_background")
```

```python
img = cv2.imread('DATA/dog_backpack.jpg')
```

```python
plt.imshow(img)
```

```python
fixed_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fixed_img)
```

#### TASK: Flip the image upside down and display it in the notebook.

```python
flipped = cv2.flip(fixed_img, 0)
plt.imshow(flipped)
```

#### TASK: Draw an empty RED rectangle around the dogs face and display the image in the notebook.

```python
fixed_img.shape
```

```python
rec = cv2.rectangle(fixed_img, pt1=(200,400), pt2=(600,700), color=(255, 0, 0), thickness=5)
plt.imshow(rec)
```

#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
# tri = cv2.line(fixed_img, pt1=(200,700), pt2=(600,700), color=(255,255, 0), thickness=10)
# tri = cv2.line(fixed_img, pt1=(200,700), pt2=(400,400), color=(255,255, 0), thickness=10)
# tri = cv2.line(fixed_img, pt1=(400,400), pt2=(600,700), color=(255,255, 0), thickness=10)
# plt.imshow(tri)

# recrate the fixed_img
fixed_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

vertices = np.array([ [250,700], [450,400], [650,700], [250, 700] ], np.int32)
pts = vertices.reshape((-1,1,2))
cv2.polylines(fixed_img, [pts], isClosed=True, color=(255,0,0), thickness=10)
plt.imshow(fixed_img)
```

### BONUS TASK. Can you figure our how to fill in this triangle? It requires a different function that we didn't show in the lecture! See if you can use google search to find it.

```python
cv2.fillPoly(fixed_img, [pts], color=(255,255,0))
plt.imshow(fixed_img)
```
