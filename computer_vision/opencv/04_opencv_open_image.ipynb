# Using OpenCV to open and handle Images

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
plt.style.use('dark_background')
```

```python
img = cv2.imread('DATA/00-puppy.jpg')
```

```python
# confirms the image is imported as a numpy array
img.shape
```

```python
# showing the image imported via openCV will
# have an unusual result typically and this 
# is because the way that matplotlib and opencv
#understand the colour channels order. 
plt.imshow(img)

# MATPLOTLIB = Red, Green, Blue
# OPENCV = BLUE, GREEN, RED
```

```python
# to correct the channel swapping we can do the following: 
fixed_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fixed_image)
```

### Exploring Color Mappings

```python
new_img = cv2.imread('DATA/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(new_img, cmap='gray')
```

### Resizing Images

```python
plt.imshow(fixed_image)
```

```python
squished = cv2.resize(fixed_image, (400, 700))
```

```python
plt.imshow(squished)
```

```python
# as a way of using a multiplier you can use the following sequence
# to resize an image to calculated measurements of the orifinal 
# source. In the example we pass 0.5 to half the size of the image
# while respecting the aspect ratio.
w_ratio = 0.5
h_ratio = 0.5
new_img = cv2.resize(fixed_image, (0,0), fixed_image, w_ratio, h_ratio)
plt.imshow(new_img)
```

### Flipping Images 

```python
new_img = cv2.flip(fixed_image, 1)
plt.imshow(new_img)
```
