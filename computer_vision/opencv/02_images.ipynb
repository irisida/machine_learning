# Working with Images and numpy

Here we will learn how to: 
- Use PIL (Python Image Library)
- How to convert an image to a numpy array with `np.asarry`
- access the shape of image arrays and separate out individual color channels

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
```

### Load image data

```python

pic = Image.open('DATA/00-puppy.jpg')
```

```python
# the type of pic is a jpg image file, this 
# is convertable to an np array with the 
# asarray(obj) method.
pic_arr = np.asarray(pic)
```

```python
type(pic_arr)
```

```python
# shape will demonstrate the height, width, depth
# you can see the 3 for depth which will match to
# the rgb channels or rgb model.
pic_arr.shape
```

### Loading and showing images as a plot 

```python
plt.style.use('dark_background')
plt.imshow(pic_arr)
```

### Assess the RGB channells as greyscale

```python
# create and show the red channel values
# values between 0-255
pic_red = pic_arr.copy()
plt.imshow(pic_red[:,:, 0], cmap='gray')
```

```python
# create and show the green channel values
# values between 0-255
pic_green = pic_arr.copy()
plt.imshow(pic_green[:,:, 1], cmap='gray')
```

```python
# create and show the blue channel values
# values between 0-255
pic_blue = pic_arr.copy()
plt.imshow(pic_blue[:,:, 2], cmap='gray')
```

For each channel we can see the greyscale from white to pure black, for each channel that scale equates to how close to pure red, green or blue a value is, 0 or pure black is for no colour for that channel found, pure white would be the max value for the color found. Per channel

### Zeroing channel values

```python
#create a copy
filtered_pic = pic_arr.copy()

# zero the green channel
filtered_pic[:,:,1] = 0

# zero the blue channel 
filtered_pic[:,:, 2] = 0

plt.imshow(filtered_pic)
```
